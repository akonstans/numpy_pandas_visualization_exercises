```python
x_rows = 2
x_columns = 2
x = np.array(range(0, 6))
a = np.sqrt(x)
b = x ** 3
c = 2 ** x
d = 1/(x+1)
```

```python
plt.plot(x, a)
plt.title('x, a')
```

```python
plt.plot(x, b)
plt.title('x, b')
```

```python
plt.plot(x, c)
plt.title('x, c')
```

```python
plt.plot(x, d)
plt.title('x, d')
```

```python
plt.subplot(x_rows, x_columns, 1)
plt.plot(x, a)
plt.title('x, a')

plt.subplot(x_rows, x_columns, 2)
plt.plot(x, b)
plt.title('x, b')

plt.subplot(x_rows, x_columns, 3)
plt.plot(x, c)
plt.title('x, c')

plt.subplot(x_rows, x_columns, 4)
plt.plot(x, d)
plt.title('x, a')
```

```python
plt.plot(x, a, c = 'cyan', label = 'sqrt(x)')
plt.plot(x, b, c = 'firebrick', label = 'x ** 3')
plt.plot(x, c, c = 'slategrey', label = '2 ** x')
plt.plot(x, d, c = 'green', label = '1/(x+1)')
plt.title('4 graphs')
plt.legend(loc = 'upper left')
plt.ylim(-1,10)
plt.show()
```
