```python
urlc = nl.get_db_url('chipotle')
```

```python
query = '''
SELECT * FROM orders
'''
```

```python
orders = pd.read_sql(query, urlc)
orders
```

```python
orders['clean_price'] = orders.item_price.apply(nl.clean)
orders
```

```python
ordata = (
    
    orders[['item_name', 'quantity', 'clean_price']]
    .groupby('item_name')
    .agg('sum')
    .sort_values('quantity', ascending = False)
    .nlargest(4, 'quantity')

)
ordata
```

```python
ordata1 = ordata.groupby('item_name').clean_price.sum().reset_index()
ordata1
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='item_name', y='clean_price', data = ordata1)
```
