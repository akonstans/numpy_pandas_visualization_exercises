```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.dtypes
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col='time', data=tips)
plt.show()
```

```python
sns.relplot(x= 'total_bill', y= 'tip', hue = 'smoker', palette= 'muted', data= tips)
plt.show()
```

```python
sns.relplot(x= 'total_bill', y= 'tip', style = 'smoker', palette= 'muted', data= tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col='time', hue='size', palette= 'muted', style='smoker', data=tips)
plt.show()
```

```python
sns.histplot(tips.total_bill)
plt.show()
```

```python
sns.histplot(tips.total_bill, kde=True)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.suptitle('Boxplots with Seaborn')

plt.subplot(221)
sns.boxplot(data=tips, y='tip')
plt.title('A "typical" boxplot')

plt.subplot(222)
sns.boxplot(data=tips, x='tip')
plt.title('Rotated 90$\degree$')

plt.subplot(223)
sns.boxplot(data=tips, y='tip', x='size')
plt.title('Providing an x and y value groups the data')

plt.subplot(224)
sns.boxplot(data=tips, y='time', x='tip', hue='smoker')
plt.title('You can also use other kwargs, like hue')

plt.subplots_adjust(hspace=0.4)
plt.show()
```

```python
data = pd.crosstab(tips.time, tips['size'])
sns.heatmap(data, annot=True, cmap=plt.cm.Greens)
data
plt.show()
```

```python
def scale(series):
    return (series - series.mean()) / series.std()

df = pd.DataFrame({'bill': scale(tips.total_bill),
                   'tip': scale(tips.tip)})

plt.subplot(211)
sns.histplot(df.bill)
plt.title('bill')

plt.subplot(212)
sns.histplot(df.tip)
plt.title('tip')
plt.show()
```
