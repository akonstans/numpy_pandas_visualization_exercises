```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
x = np.array(range(1,21))
y = (x ** x)
z = list(range(1, 21))
m = [n ** n for n in z]
```

```python
a = 0*x+1
b = np.log10(x)
c = x
d = x * np.log10(x)
e = x ** 2
f = 2 ** x
g = [math.factorial(n) for n in x]
h = [n ** n for n in z]
i = [n ** n for n in x]
```

```python
plt.plot(x, a, label = 'O(1)')
plt.plot(x, b, label = 'O (log n)')
plt.plot(x, c, label = 'O(n)')
plt.plot(x, d, label = 'O(n log n)')
plt.plot(x, e, label = 'O(n^2)')
plt.plot(x, f, label = 'O(2^n)')
plt.plot(x, g, label = 'O(n!)')
plt.plot(z, h, label = 'O(n^n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc = 'upper left')
plt.ylim(0, 10000)
```
